     ***************************** Webscraping for City - Dubai *****************************

```python
driver.get('https://www.booking.com/searchresults.en-gb.html?label=yho748jc-1FCAEoggI46AdIM1gDaGyIAQGYAQm4ARfIAQzYAQHoAQH4AQyIAgGoAgO4AurS96YGwAIB0gIkMTJhN2VkNjAtYzc1MS00MjVhLTg3ZjYtMzI5ZWMzNzlkNzhm2AIG4AIB&sid=b8f957b39c045a637fc32a3a001d95d0&aid=397646&ss=Dubai&ssne=Dubai&ssne_untouched=Dubai&efdco=1&lang=en-gb&src=searchresults&dest_id=-782831&dest_type=city&checkin=2023-10-21&checkout=2023-10-22&ltfd=1%3A1%3A11-2023%3A1%3A1&group_adults=2&no_rooms=1&group_children=0&nflt=ht_id%3D204')
```

```python
import pandas as pd


hotel_data = []

# Define the number of pages you want to scrape (5 in this case)
start_page = 1
end_page = 25

for page in range(start_page, end_page + 1):
    try:
        # Extract data using XPaths
        hotel_name_elements = driver.find_elements(By.XPATH, '//div[@class="f6431b446c a15b38c233"]')
        room_type_elements = driver.find_elements(By.XPATH, '//span[@class="e8f7c070a7"]')
        single_bed_elements = driver.find_elements(By.XPATH, '//div[@class="fc367255e6"]')
        double_bed_elements = driver.find_elements(By.XPATH, '//div[@class="fc367255e6"]')
        price_elements = driver.find_elements(By.XPATH, '//span[@class="f6431b446c fbfd7c1165 e84eb96b1f"]')
        
        num_elements = min(len(hotel_name_elements), len(room_type_elements), len(single_bed_elements), len(double_bed_elements), len(price_elements))
        
        for i in range(num_elements):
            hotel_name = hotel_name_elements[i].text
            room_type = room_type_elements[i].text if i < len(room_type_elements) else "N/A"
            
            # Check if the extracted text contains "single" and "double"
            single_bed_text = single_bed_elements[i].text if i < len(single_bed_elements) else "N/A"
            double_bed_text = double_bed_elements[i].text if i < len(double_bed_elements) else "N/A"
            
            single_bed = "Yes" if "single" in single_bed_text.lower() else "No"
            double_bed = "Yes" if "double" in double_bed_text.lower() else "No"
            
            price = price_elements[i].text if i < len(price_elements) else "N/A"

            # Store the data in a suitable data structure, e.g., a list of dictionaries
            hotel_data.append({
                "Hotel Name": hotel_name,
                "Room Type": room_type,
                "Single Bed": single_bed,
                "Double Bed": double_bed,
                "Price": price
            })
    except NoSuchElementException:
        # Handle the case when an element is not found
        hotel_data.append({
            "Hotel Name": "N/A",
            "Room Type": "N/A",
            "Single Bed": "N/A",
            "Double Bed": "N/A",
            "Price": "N/A"
        })
    
    try:
        # Click the "Next" button to navigate to the next page
        nxt_button = driver.find_element(By.XPATH, '//div[@class="b16a89683f cab1524053"]')
        nxt_button.click()
    except NoSuchElementException:
        # Handle the case when the "Next" button is not found
        print("Next button not found. Exiting the loop.")
        break
    
    # Add a short delay to allow the next page to load
    time.sleep(2)

# Create a DataFrame from the scraped data
df_Dubai = pd.DataFrame(hotel_data)

# Print the DataFrame
print(df_Dubai)
```
