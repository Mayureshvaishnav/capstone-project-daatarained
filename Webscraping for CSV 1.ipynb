                                *********Scraping for Dubai city************

```python
driver.get('https://www.booking.com/searchresults.en-gb.html?label=yho748jc-1FCAEoggI46AdIM1gDaGyIAQGYAQm4ARfIAQzYAQHoAQH4AQyIAgGoAgO4AurS96YGwAIB0gIkMTJhN2VkNjAtYzc1MS00MjVhLTg3ZjYtMzI5ZWMzNzlkNzhm2AIG4AIB&sid=b8f957b39c045a637fc32a3a001d95d0&aid=397646&city=-782831&nflt=ht_id%3D204&offset=0')
```

```python
import pandas as pd
import time

hotel_reviews = []

# Define the number of pages you want to scrape (5 in this case)
start_page = 1
end_page = 9

for page in range(start_page, end_page + 1):
    # Add your scraping code here (similar to the previous code)
    review_elements = driver.find_elements(By.XPATH, '//div[@data-testid="review-score"]//div[@aria-label]')
    reviews_elements = driver.find_elements(By.XPATH, '//div[@data-testid="review-score"]//div[@class="abf093bdfe f45d8e4c32 d935416c47"]')
    score_elements = driver.find_elements(By.XPATH, '//div[@data-testid="review-score"]//div[contains(@aria-label, "Scored")]')
    name_elements = driver.find_elements(By.XPATH, '//div[@data-testid="title"]')  # Add XPath for hotel names
    city_elements = driver.find_elements(By.XPATH, '//span[@class="aee5343fdb def9bc142a"]')  # Add XPath for city
    
    for review, reviews, score, name, city in zip(review_elements, reviews_elements, score_elements, name_elements, city_elements):
        review_text = review.text.strip()
        reviews_text = reviews.text.strip()
        score_text = score.text.strip()
        name_text = name.text.strip()  # Extract hotel names
        city_text = city.text.strip()  # Extract city
        
        # Store the data in a suitable data structure, e.g., a list of dictionaries
        hotel_reviews.append({
            "Hotel Name": name_text,
            "City": city_text,  # Include city information
            "Review": review_text,
            "Number of Reviews": reviews_text,
            "Review Score": score_text
        })

    # Click the "Next" button to navigate to the next page
    nxt_button = driver.find_element(By.XPATH, '//div[@class="b16a89683f cab1524053"]')
    nxt_button.click()
    
    # Add a short delay to allow the next page to load
    time.sleep(2)

# Create a DataFrame from the scraped data
df_dubai = pd.DataFrame(hotel_reviews)

# Print the DataFrame
print(df_dubai)
```

```python
df_dubai['City'] = 'Dubai'
```

```python
df_dubai['Number of Reviews'] = df_dubai['Number of Reviews'].str.replace('[^\d]', '', regex=True)
```
