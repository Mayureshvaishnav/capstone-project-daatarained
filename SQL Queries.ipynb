```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('mydatabase2.db')

query = """
    SELECT `Breakfast Cuisine 1` AS Cuisine, COUNT(`Breakfast Cuisine 1`) AS Count
    FROM mytable2
    GROUP BY `Breakfast Cuisine 1`
    
    UNION ALL
    
    SELECT `Breakfast Cuisine 2` AS Cuisine, COUNT(`Breakfast Cuisine 2`) AS Count
    FROM mytable2
    GROUP BY `Breakfast Cuisine 2`
    
    UNION ALL
    
    SELECT `Breakfast Cuisine 3` AS Cuisine, COUNT(`Breakfast Cuisine 3`) AS Count
    FROM mytable2
    GROUP BY `Breakfast Cuisine 3`"""

df = pd.read_sql_query(query, conn)

df
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('mydatabase3.db')

query = """
SELECT *
FROM mytable3
WHERE "Double Bed" = 'Yes'
ORDER BY CAST(REPLACE(Price, '₹ ', '') AS DECIMAL) DESC;
"""

df = pd.read_sql_query(query, conn)

df
```

                                                    Query 2

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('mydatabase3.db')

query = """SELECT "Room Type", 
       AVG(CAST(REPLACE(SUBSTR(Price, 2), ',', '') AS DECIMAL)) AS Average_Price
FROM mytable3
GROUP BY "Room Type";
"""

df = pd.read_sql_query(query, conn)

df
```

```python
import sqlite3
import pandas as pd

df1 = pd.read_csv('csv1.csv')
df2 = pd.read_csv('csv2.csv')

conn = sqlite3.connect('my_combined_database.db')  # You can use any name for your combined database

df1.to_sql('table1', conn, if_exists='replace', index=False)
df2.to_sql('table2', conn, if_exists='replace', index=False)
```

                                                    Subquery 1

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('my_combined_database.db')

query = """
SELECT table1."Hotel Name", table2."Free WiFi"
FROM table1
JOIN table2 ON table1."Hotel Name" = table2."Hotel Name"
WHERE table1."Review Score" > 5 AND table2."Free WiFi" = 'yes';
"""

df = pd.read_sql_query(query, conn)

df
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('my_combined_database.db')

query = """
SELECT table1."Hotel Name", table2."Family rooms", table2."Non-smoking rooms"
FROM table1
JOIN table2 ON table1."Hotel Name" = table2."Hotel Name"
WHERE table2."Family rooms" = 'yes' AND table2."Non-smoking rooms" = 'yes';
"""

df = pd.read_sql_query(query, conn)

df
```

```python
import sqlite3
import pandas as pd

df1 = pd.read_csv('csv2.csv')
df2 = pd.read_csv('csv3.csv')

conn = sqlite3.connect('my_combined_database1.db')

df1.to_sql('table2', conn, if_exists='replace', index=False)
df2.to_sql('table3', conn, if_exists='replace', index=False)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('my_combined_database1.db')

query = """
SELECT table3."Room Type", 
       ROUND(AVG(CAST(REPLACE(table3."Price", '₹', '') AS DECIMAL)) , 0) AS Average_Price
FROM table3
JOIN table2 ON table3."Hotel Name" = table2."Hotel Name"
GROUP BY table3."Room Type";
"""

df = pd.read_sql_query(query, conn)

df
```
